```python
import numpy as np 
import matplotlib.pyplot as plt
import random

#Assigning constants as given in the problem. 

x0 = 1
lam = 0.5          #lambda
q = 1 
del_t = 0.01
n_itr = 101      #number of trajectories
num_points = 101   #number of points to be plotted

time_vec = [i*del_t for i in range(num_points)]
#print(len(time_vec),time_vec)
 
for n in range(n_itr) : 
    x_vec = [0] * num_points      #initializing x vector to Zero vector
    x_vec[0] = x0                 #Assigning first element as x0
    for i in range(num_points-1) :
        #print(i)
        if n == 0 : 
            del_bk = 0            #Assigning gaussian var = 0 for Zero noise solution for mean trajectory
        else : 
            del_bk = np.random.normal(0, np.sqrt(q * del_t))     #defining gaussian random variable
        x_vec[i+1] = x_vec[i] - 0.5 * x_vec[i] * del_t + del_bk 
    #print(x_vec)
    
    if n == 0 :            
        x_mean = x_vec      #saving the x_vec corr to mean trajectory
    else :
        pass 
    
    #Calculating x_vec corr to Variance trajectory
    var_list = [0] * num_points        
    diff_list = [vec_el - mean_el for vec_el, mean_el in zip(x_vec,x_mean)]          #(xi -x)
    var_list_i = [item **2 / (num_points - 1) for item in diff_list]                 #(xi-x)^2 / (n-1)
    var_list = [var_el + var_el_i for var_el, var_el_i in zip(var_list,var_list_i)]  #sum((xi-x)^2 / (n-1))
    #print(var_list)
    plt.plot(time_vec, x_vec, color = 'b')         #plotting all the trajectories in blue
    #print(x_vec)
#print(var_list)
plt.plot(time_vec, var_list, color = 'r')          #plotting variance trajectory in red
plt.plot(time_vec,x_mean, color = 'k')             #plotting mean trajectory in black

plt.xlabel("time")
plt.ylabel("x(t)")

plt.show()
```
